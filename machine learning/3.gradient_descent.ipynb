```python
x = np.arange(0,100,1)
fake_noise = 15*np.random.randn(1,100)  + 25
y = 3*x  + fake_noise
```

```python
def cost_function(w):
    h = w*x
    MSE =np.mean((y - h)**2)
    return MSE
cost_function(3)
```

```python
weight = np.arange(2,4,0.2)
for w in weight:
    h = w*x #Trend line
    plt.plot(x,h,color = "pink")
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
import numpy as np

def gradient_descent(x, y):
    m = b = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.001
    for i in range(iterations):
        # Calculate predicted value
        h = m * x + b
        
        # Calculate slope for m and b
        m_derivative = -sum(x * (y - h))
        b_derivative = -sum(y - h)
        
        # Update new value for m and b
        m = m - learning_rate * m_derivative
        b = b - learning_rate * b_derivative
        
        # calculate cost
        cost = (1/2) * sum([val ** 2 for val in (y - h)])
        plt.scatter(i,cost,c = "red")
        #print("m {}, b {}, iteration {}, cost {}".format(m, b, i, cost ))

x = np.array([1,2,3,4,5])
y = np.array([5, 7, 9, 11, 13])

gradient_descent(x, y)
```
