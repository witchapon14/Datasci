```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/Fake News Detection Dataset.csv")
df
```

```python
df.describe()
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Word_Count'], kde=True, color='blue', bins=30)
plt.title('Distribution of Word Count')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Number_of_Sentence'], kde=True, color='red', bins=30)
plt.title('Distribution of Number of Sentences')
plt.xlabel('Number of Sentences')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Label', y='Word_Count', data=df)
plt.title('Boxplot of Word Count by Label')
plt.xlabel('Label')
plt.ylabel('Word Count')
plt.show()
```
